```python
search_words = ["nfts OR nft OR NFTCommunity OR NFTCollection OR nftart OR NFTProject OR NFTMarketplace OR NFTGiveaway OR opensea"]
date = "2022-05-31"
# geo = "-6.14591,106.72772,3000km"

header = ['waktu','username','tweet']
with open('data/CrawlingDataNft.csv', 'w', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    
    for tweet in tweepy.Cursor(api.search_tweets,q=search_words, until=date, count=100, lang="id").items(1000):
        print(tweet.created_at, tweet.user.screen_name, tweet.text)
        twitt = ([tweet.created_at, tweet.user.screen_name, tweet.text])
        writer.writerow(twitt)
```
